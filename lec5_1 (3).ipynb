```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
Me=0.51
h=1
c=1
E=v=0.5   
epsilon0=(Me*c*c)/((Me*c*c)+(2*h*v))
gamma=(h*v)/(Me*c*c)

alpha1=np.log(1/epsilon0)
alpha2=np.log(1-epsilon0**2)/2

alpha1_list=[]
alpha2_list=[]
epsilon_list=[]
sin_Theta_square_list=[]
g_list=[]
phi_list=[]
Photon_energy_list=[]
theta_list=[]
phi_list=[]
r0_list=[]
r1_list=[]
r2_list=[]
E_energy_list=[]

for i in range(10000):
    r0=random.uniform(0,1)
    r1=random.uniform(0,1)
    r2=random.uniform(0,1)

    if r0<((alpha1)/(alpha1+alpha2)):
        epsilon=epsilon0**r1
        f1=1/(alpha1*epsilon)

    else:
        epsilon=np.sqrt(epsilon0**2+(1-epsilon0**2)*r1)
        f2=epsilon/alpha2

    t=Me*c*c*(1-epsilon)/(h*v*epsilon)
    sin_Theta_square=t*(2-t)
    theta=np.arcsin(np.sqrt(t*(2-t)))
    phi=np.arctan(1/((1+E/Me)*np.tan(theta/2)))

    g=(1-(epsilon/(1+epsilon**2))*sin_Theta_square)

    T=E-epsilon*E

    if g>=r2:
        epsilon_list.append(epsilon)
        theta_list.append(theta)
        phi_list.append(phi)
        r0_list.append(r0)
        r1_list.append(r1)
        r2_list.append(r2)
        alpha1_list.append(alpha1)
        alpha2_list.append(alpha2)
        epsilon_list.append(epsilon)
        sin_Theta_square_list.append(sin_Theta_square)
        g_list.append(g)
        Photon_energy_list.append(epsilon*E)
        E_energy_list.append(T)

    
    
```

```python
plt.figure(figsize=(12,9))

plt.subplot(2,2,1)
plt.hist(E_energy_list , bins=100,color='blue')
plt.title('Energy of electron')
plt.xlabel('Energy of electron')
plt.ylabel('Relative intensity')



plt.subplot(2,2,2)
plt.hist(Photon_energy_list , bins=100,color='blue')
plt.title('Energy of photon')
plt.xlabel('Energy of photon')
plt.ylabel('Relative intensity')


plt.subplot(2,2,3)
plt.hist(phi_list , bins=100,color='blue')
plt.title('phi')
plt.xlabel('phi')
plt.ylabel('Relative intensity')


plt.subplot(2,2,4)
plt.hist(theta_list , bins=100,color='blue')
plt.title('Theta')
plt.xlabel('Theta')
plt.ylabel('Relative intensity')


plt.show()
```
